```python
from umap import UMAP
import seaborn as sns
from gensim.models import Word2Vec, KeyedVectors
import numpy as np
```

```python
model = KeyedVectors.load("models/lower_streets_model", mmap='r')
```

```python
vectors = []
for word in model.wv.vocab:
    vectors.append(model[word])

vectors = np.array(vectors)
```

```python
reducer = UMAP(verbose=True)
embeddings = reducer.fit_transform(vectors)
```

```python
import numpy as np

from bokeh.models import ColumnDataSource, Plot, LinearAxis, Grid
from bokeh.models.glyphs import Text
from bokeh.io import show
from bokeh.plotting import output_notebook, figure
output_notebook()

x = embeddings[:, 0]
y = embeddings[:, 1]
text = list(model.wv.vocab.keys())

source = ColumnDataSource(dict(x=x, y=y, text=text))

plot = figure()

glyph = Text(x="x", y="y", text="text", text_color="lightblue")
plot.add_glyph(source, glyph)

show(plot)
```
